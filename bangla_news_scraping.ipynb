```python
def tags_processing(modified_dict_data_tags):
  unprocessed_tag_list = modified_dict_data_tags
  processed_tag_list = []
  processed_tag_dict = {}
  for unprocessed_tag in unprocessed_tag_list:
    tag_name = unprocessed_tag.get('name', None)
    tag_properties = unprocessed_tag.get('properties', None)
    tag_description = tag_properties.get('tag-description', None)
    processed_tag_dict.update({'name': tag_name})
    meta_keywords = tag_properties.get('meta-keywords', None)
    processed_tag_dict.update({'tag-description': tag_description})
    processed_tag_dict.update({'tag-meta-keywords': meta_keywords})
    processed_tag_list.append(processed_tag_dict)
  return processed_tag_list
```

```python
def list_checking(data):
  length = len(data)
  if length > 0:
    return data[0]
  else:
    return data
```

```python
def processing_data_from_api(modified_dict_data_items):

  modified_dict_data_list = []

  for item in modified_dict_data_items:
    news_col_name = f"news_details_headline_item"
    item_single = item.get('item', None)
    item_headline = item_single.get('headline', None)
    temp_dict_data = {}
    temp_dict_data[f"{news_col_name}_headline"] = list_checking(item_headline)
    item_single_story = item.get('story', None)
    item_headline_seo = item_single_story.get('seo', None)
    modified_dict_data_tags = item_single_story.get('tags', None)
    item_headline_author_name = item_single_story.get('author-name', None)
    item_headline_meta_keywords = item_headline_seo.get('meta-keywords', None)
    item_headline_meta_description = item_headline_seo.get('meta-description', None)
    temp_dict_data[f"{news_col_name}_meta-keywords"] = item_headline_meta_keywords
    temp_dict_data[f"{news_col_name}_meta-description"] = item_headline_meta_description
    temp_dict_data[f"{news_col_name}_author-name"] = item_headline_author_name
    tag_lists = tags_processing(modified_dict_data_tags)
    temp_dict_data[f"{news_col_name}_tag_list"] = tag_lists
    item_story = item.get('story', None)
    item_story_alternative = item.get('alternative', None)
    item_story_alternative_home = item.get('home', None)
    item_story_alternative_home_default = item.get('default', None)
    item_story_alternative_home_default_hero_caption = item.get('hero-image-caption', None)
    temp_dict_data[f"{news_col_name}_image_caption"] = item_story_alternative_home_default_hero_caption
    modified_dict_data_list.append(temp_dict_data)
  
  return modified_dict_data_list
```
